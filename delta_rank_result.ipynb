In this session, we are going to see the post-process result of covid delta mutation escape mutant priority rank.

```python
import pandas as pd
file_name = "extended_analyze_semantics_delta_bilstm_512.csv"
delta_df = pd.read_csv(file_name, sep='\t')
delta_df
```

```python
# Mark is_escape label manually
escape_df = delta_df[delta_df['is_escape'] == True]
escape_df
```

```python
"""Make rank of 'prob', 'change', 'codon indep'.
"""
rank_names = ['grammar rank', 'sem change rank', 'indep rank']
score_names = ['prob', 'change', 'codon indep']
def rank_helper(df, rank_names, score_names):
    df[rank_names] = df[score_names].rank(method="min", ascending=False)

rank_helper(delta_df, rank_names, score_names)
delta_df
```

```python
from itertools import combinations
"""make combined rank"""
def combined_rank_helper(df, rank_names, num):
    for combined_rank in combinations(rank_names, num):
        df['+'.join(combined_rank) + " rank"] = (df[list(combined_rank)].sum(axis=1)).rank(method='min')
combined_rank_helper(delta_df, rank_names, 1)
combined_rank_helper(delta_df, rank_names, 2)
combined_rank_helper(delta_df, rank_names, 3)
delta_df
```

```python
def mean_std_helper(df, name):
    return f'mean and std of {name} is\t{df[name].mean()},\t{df[name].std()}'

delta_escape_df = delta_df[delta_df['is_escape'] == True]

with open('delta_escape_result.txt', 'w') as file:
    for text in map(lambda x: mean_std_helper(delta_escape_df, x), filter(lambda string: "rank rank" in string, delta_escape_df.columns)):
        print(text)
        file.write(text + '\n')
```
