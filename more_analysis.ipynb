2. proper ratio(coefficients) between grammarticality and codon mutant prob
3. seeing result difference after applying proper coeffs rbd domain
4. applying codon mutation algorithm with another reference and check the difference again

```python
# finding out proper ratio with simple method
# from now on we use only independent probabilities
# requires: 2 cases of extended ranked table. one from reference, the other from our algorithm
# basic codes were imported from 'regularized_cscs.ipynb'
# For the normalization, instead of using 'rank', we are using log of probabilities.

# but for comparison, we will use both rank and log(prob) as well.

# recreate extended table. The extended table will include indep value from 
# both reference and our algorithm.
import pandas as pd
import numpy as np
df = pd.read_csv('postprocess_results/ranked_extended_analyze_semantics_cov_bilstm_512.txt', sep='\t')
# add normalized log(prob) & change
df['log_prob'] = np.log(df['prob'])
df

```

```python
def normalize_series(series, method):
    return method(series)

def min_max_normalize(series):
    return (series - series.min()) / (series.max() - series.min())

def mean_std_normalize(series):
    return (series - series.mean()) / series.std()

df['min_max_log_prob'] = normalize_series(df['log_prob'], min_max_normalize)
df['min_max_change'] = normalize_series(df['change'], min_max_normalize)
df['mean_std_log_prob'] = normalize_series(df['log_prob'], mean_std_normalize)
df['mean_std_change'] = normalize_series(df['change'], mean_std_normalize)
df
```

```python
"""Investigate 1
we will find out proper coefficients for combining grammar and sem change both for rank
and normalized log(prob)
"""
"""cols to investigate & desired name of result: 
1. grammar rank, semantic chagne rank, sum_rank
2. min_max_log_prob, min_max_change, sum_min_max
3. mean_std_log_prob, mean_std_change, sum_mean_std
"""
import sys
def investigate(gram_name, sem_name, sum_rank_name, inc=0.1):
    coef = 0
    min_val = (0, sys.maxsize)
    for _ in range(0, 10):
        coef += inc
        df[sum_rank_name] = (df[gram_name] + coef * df[sem_name]).rank(ascending=False)
        df_escape = df[df['is_escape']]
        current_val = df_escape[sum_rank_name].mean()
        min_val = (coef, current_val) if min_val[1] > current_val else min_val
    return min_val
sum_rank_result = investigate('grammar rank', 'semantic change rank', 'sum_rank')
min_max_result = investigate('min_max_log_prob', 'min_max_change', 'sum_min_max')
mean_std_result = investigate('mean_std_log_prob', 'mean_std_change', 'sum_mean_std')

print(f'sum rank result = {sum_rank_result}')
print(f'min_max_norm result = {min_max_result}')
print(f'mean_std_norm result= {mean_std_result}')
```

```python
"""second investigate
find good number for 3 combined values: grammar, semantic change, codon indep
sum rank, min_max, mean_std will be used
"""
def investigate3(gram_name, sem_name, codon_name, sum_rank_name):
    min_val = (0, 0, sys.maxsize)
    for i in np.linspace(0.1, 10, 100):
        for j in np.linspace(0.1, 1, 100):
            df[sum_rank_name] = (j * df[gram_name] + (1 - j) * df[codon_name] + i * df[sem_name]).rank(ascending=False)
            df_escape = df[df['is_escape']]
            current_val = df_escape[sum_rank_name].mean()
            min_val = (i, j, current_val) if min_val[2] > current_val else min_val
    return min_val

df['log_codon_prob'] = np.log(df['codon indep'])
df['mean_std_log_codon_prob'] = normalize_series(df['log_codon_prob'], mean_std_normalize)
df['min_max_log_codon_prob'] = normalize_series(df['log_codon_prob'], min_max_normalize)

df['indep rank'] = df['codon indep'].rank(ascending=True)

sum_rank_result = investigate3('grammar rank', 'semantic change rank', 'indep rank', 'sum_rank')
min_max_result = investigate3('min_max_log_prob', 'min_max_change', 'min_max_log_codon_prob', 'sum_min_max')
mean_std_result = investigate3('mean_std_log_prob', 'mean_std_change', 'mean_std_log_codon_prob', 'sum_mean_std')

print(f'sum rank result(gram, codon, sem, mean) = {sum_rank_result[1], 1-sum_rank_result[1], sum_rank_result[0], sum_rank_result[2]}')
print(f'min_max_norm result(gram, codon, sem, mean)= {min_max_result[1], 1-min_max_result[1], min_max_result[0], min_max_result[2]}')
print(f'mean_std_norm result(gram, codon, sem, mean)= {mean_std_result[1], 1-mean_std_result[1], mean_std_result[0], mean_std_result[2]}')
```
