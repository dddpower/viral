In this session we are going to generate combined ranks and get mean and standard deviation of the combined ranks of covid, flu, hiv.

```python
import pandas as pd
# fnames = ["extended_analyze_semantics_cov_bilstm_512.txt",
#           "extended_analyze_semantics_flu_h1_bilstm_512.txt",
#           "extended_analyze_semantics_hiv_bilstm_512.txt"]
# cov_df, flu_df, hiv_df = tuple(map(lambda x: pd.read_csv(x, sep='\t'), fnames))
cov_df = pd.read_csv("extended_analyze_semantics_cov2rbd_bilstm_512.txt", sep='\t')
cov_df
```

```python
"""Make rank of 'prob', 'change', 'codon indep', 'codon dep'.
"""
rank_names = ['grammar rank', 'sem change rank', 'indep rank', 'dep rank']
score_names = ['prob', 'change', 'codon indep', 'codon dep']
rank_names_hiv = ['grammar rank', 'sem change rank', 'dep rank']
score_names_hiv = ['prob', 'change', 'codon dep']
def rank_helper(df, rank_names, score_names):
    df[rank_names] = df[score_names].rank(method="min", ascending=False)

rank_helper(cov_df, rank_names, score_names)
# rank_helper(flu_df, rank_names, score_names)
# rank_helper(hiv_df, rank_names_hiv, score_names_hiv)
# # hiv doesn't have 'indep rank'
# hiv_df
```

```python
from itertools import combinations
"""make combined rank"""
def combined_rank_helper(df, rank_names, num):
    for combined_rank in combinations(rank_names, num):
        df['+'.join(combined_rank) + " rank"] = (df[list(combined_rank)].sum(axis=1)).rank(method='min')
combined_rank_helper(cov_df, rank_names, 1)
combined_rank_helper(cov_df, rank_names, 2)
combined_rank_helper(cov_df, rank_names, 3)

combined_rank_helper(flu_df, rank_names, 1)
combined_rank_helper(flu_df, rank_names, 2)
combined_rank_helper(flu_df, rank_names, 3)

combined_rank_helper(hiv_df, rank_names_hiv, 1)
combined_rank_helper(hiv_df, rank_names_hiv, 2)
combined_rank_helper(hiv_df, rank_names_hiv, 3)
cov_df
```

```python
def mean_std_helper(df, name):
    return f'mean and std of {name} is\t{df[name].mean()},\t{df[name].std()}'

def escape_filter(df):
    return df[df['is_escape'] == True]
cov_df_escape, flu_df_escape, hiv_df_escape = map(escape_filter, (cov_df, flu_df, hiv_df))


for df, file_name in ((cov_df_escape, "cov_escape_result.txt"), (flu_df_escape, "flu_escape_result.txt"), (hiv_df_escape, 'hiv_escape_result.txt')):
    with open(file_name, 'w') as file:
        for text in map(lambda x: mean_std_helper(df, x), filter(lambda string: "rank rank" in string, df.columns)):
            print(text)
            file.write(text + '\n')
```

```python
cov_df_escape.to_csv("cov_escape.csv")
flu_df_escape.to_csv("flu_escape.csv")
hiv_df_escape.to_csv("hiv_escape.csv")
```
